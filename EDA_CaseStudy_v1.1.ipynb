```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
%matplotlib inline  

# Reading data file
# Since the data file had too many columns , 111 to be precise, so it was not feasible to declare the data type of each column. 
# So, we prefer keeping low_memory = False while reading.

loan_data = pd.read_csv("loan.csv",low_memory=False)
```

```python
# ELIMINATION ACTION 1:We are trying to get rid of thise columns where the missing data percentage is 70 or more.
# This decision of keeping this threshold to 70% is subjective.
seventy_percent_missing = len(loan_data)*(0.7)
column_dropped_loan_data_SET1 = loan_data.dropna(thresh=seventy_percent_missing,axis=1)
```

```python
# Checking the dimension of the dataframe post certain columns on the basis of above condition is being eliminated(ELIMINATION ACTION 1)
column_dropped_loan_data_SET1.shape
```

```python
# Checking the entire dataframe post ELIMINATION ACTION 1
column_dropped_loan_data_SET1
```

```python
# ELIMINATION ACTION 2:
# Looking into the dataframe column_dropped_loan_data_SET1 we realized there is need for further cleaning.
# There are several columns whose values only makes sense post a certain loan has already been disbursed.
# Since in this particular problem we are trying to pull out only those columns to act as the list of Predictor variables or Decision variables,
# by which we mean they shall help to decide whether at all we should disburse the loan or not,
# hence there is no need to keep those columns where certain value shall get populated only after the Loan is being disbursed for 
# with respect to the problem at hand they can be regarded as "Futuristic" and hence can lead to "Data Leakage" which can lead the ultimate model we build to "overfit".
# Below the list of columns which are "Futuristic" wrt the problem at hand. 

to_be_dropped_set1 = ['funded_amnt','funded_amnt_inv',
             'out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d','collections_12_mths_ex_med','delinq_amnt']
```

```python
# ELIMINATION ACTION 2. Dropping the "Futuristic" columns.
column_dropped_loan_data_SET2 = column_dropped_loan_data_SET1.drop(to_be_dropped_set1,axis=1)
```

```python
# Checking the dimension of the dataframe post ELININATION ACTION 2
column_dropped_loan_data_SET2.shape
```

```python
# Checking the precise list of columns post ELIMINATION ACTION 2
column_dropped_loan_data_SET2.columns
```

```python
# Checking the "Purpose" and "Title" column to decide upon which one to keep for it seems one can be safely eliminated
for name in ['purpose','title']:
    print("Unique Values in column: {}\n".format(name))
    print(column_dropped_loan_data_SET2[name].value_counts(),'\n')
```

```python
#ELIMINATION ACTION 3
# It appears the purpose and title columns do contain overlapping information "Sense-wise,
# but the purpose column contains fewer discrete values and is cleaner, so we will keep it and drop title.
column_dropped_loan_data_SET3 = column_dropped_loan_data_SET2.drop('title',axis=1)
```

```python
#ELIMINATION ACTION 4
# We are now looking for any columns that contain only one unique value and remove them. 
# These columns won't be useful for the model building since they don't add any important information and hence can be considered redundant.
# We make use of the pandas Series method nunique() since it returns the number of unique values, excluding any null values. 
column_dropped_loan_data_SET4 = column_dropped_loan_data_SET3.loc[:,column_dropped_loan_data_SET3.apply(pd.Series.nunique) != 1]
```

```python
# Checking the dimension of the data frame post ELIMINATION ACTION 4
column_dropped_loan_data_SET4.shape
```
